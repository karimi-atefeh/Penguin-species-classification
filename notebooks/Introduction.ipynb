# **Classify Penguins Species Project**

**Artwork by @allison_horst**

## <font color="orange">**Introduction**</font>


In this project, we delve into the fascinating world of penguins to develop a machine learning model capable of accurately classifying species based on the Palmers Penguins dataset. Our approach embraces the Cross-Industry Standard Process for Data Mining (CRISP-DM) methodology, ensuring a structured and systematic workflow that enhances both the development and analysis phases.

We've used the Palmers Penguins dataset, which includes three distinct penguin species: Adelie, Chinstrap, and Gentoo, which we've visually distinguished in our analysis using a color-coding scheme:
-  **Chinstrap**: Purple
-  **Gentoo**: Green
-  **Adelie**: Orange

Each color represents a unique species, aiding in a clearer visualization and interpretation of our findings. This visual distinction is not only helpful for analysis but also makes our results more accessible and understandable.

Utilizing the CRISP-DM model ensures that each step of our project, from data understanding to model evaluation, is meticulously planned and executed. This structured approach helps us navigate through complex data analysis with precision, aligning our strategies with the project's objectives efficiently.
